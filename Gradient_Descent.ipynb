gradient descent for linear regression using SciPy while ensuring clear visibility of each update step. The script includes:

Iterative updates of m and b
A stopping condition based on the change in cost function
Computation of final predictions
A visualization of how m and b evolve over iterations

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Data points
X = np.array([1, 3])
Y = np.array([3, 6])

# Learning rate and initialization
alpha = 0.1
m, b = -1, 1  # Initial guess
tolerance = 1e-6  # Stopping condition (small change in cost function)
max_iterations = 1000  # Prevent infinite loops

# Gradient descent updates
m_values, b_values = [m], [b]  # Track changes for visualization
cost_values = []  # Track cost function values

def compute_cost(m, b):
    predictions = m * X + b
    return np.mean((Y - predictions) ** 2)  # Mean Squared Error (MSE)

for iteration in range(max_iterations):
    predictions = m * X + b
    error = Y - predictions

    # Compute gradients
    gradient_m = -np.mean(error * X)
    gradient_b = -np.mean(error)

    # Update parameters
    new_m = m - alpha * gradient_m
    new_b = b - alpha * gradient_b

    # Compute new cost
    new_cost = compute_cost(new_m, new_b)

    # Store values for plotting
    m_values.append(new_m)
    b_values.append(new_b)
    cost_values.append(new_cost)

    # Stopping condition
    if abs(compute_cost(m, b) - new_cost) < tolerance:
        break

    m, b = new_m, new_b  # Update for next iteration

# Final predictions
final_predictions = m * X + b

# Visualization
plt.figure(figsize=(10, 4))

# Plot m and b changes over iterations
plt.subplot(1, 2, 1)
plt.plot(m_values, label="m (slope)")
plt.plot(b_values, label="b (intercept)")
plt.xlabel("Iteration")
plt.ylabel("Value")
plt.title("Change in m and b over iterations")
plt.legend()

# Plot cost function changes
plt.subplot(1, 2, 2)
plt.plot(cost_values, label="Cost function")
plt.xlabel("Iteration")
plt.ylabel("Cost")
plt.title("Cost function decrease")
plt.legend()

plt.tight_layout()
plt.show()

# Print final equation
print(f"Final equation: y = {m:.4f}x + {b:.4f}")
print(f"Final predictions for given points: {final_predictions}")
```

Insights:
The initial updates are large, then they decrease as we approach convergence.
The cost function graph shows a rapid initial decrease, then flattens as the model converges.
The stopping condition prevents unnecessary iterations, optimizing computational efficiency

We start with m = -1 and b = 1, a learning rate of 0.1, and a tolerance level (1e-6) to stop when the cost function changes minimally.
